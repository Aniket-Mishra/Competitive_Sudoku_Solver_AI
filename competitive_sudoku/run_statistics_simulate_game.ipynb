```python
import subprocess
import os

# Define parameters
time_inputs = [0.1, 0.5, 1.0, 5.0]
main_player = "team03_A3_best_performing"
opponents = [
    "A3_MCMC",
    "A3_MCTS_with_minimax",
    "greedy_player",
    "team03_A2",
]  # Opponents
boards = [
    "2x2",
    "2x3",
    "3x3",
    #   "3x4",
    #   "4x4"
]
results_file = "output/results_A3.csv"

# Verify that results.csv exists
if not os.path.exists(results_file):
    print(
        f"Error: '{results_file}' not found. Ensure simulate_game.py initializes this file."
    )
else:
    print(f"Using existing '{results_file}' for results.")

for turn in range(0, 15):
    for time_inp in time_inputs:
        # for opponent in opponents:
        # for board in boards:
        #     board_file = f"boards/empty-{board}.txt"
        for board in boards:
            board_file = f"boards/empty-{board}.txt"
            for opponent in opponents:

                # Compare A2_Heuristics as the first and second player
                for first, second in [
                    (main_player, opponent),
                    (opponent, main_player),
                ]:
                    # Build the command
                    command = [
                        "python3",  # Or just "python" if using Windows
                        "statistics_simulate_game.py",
                        f"--first={first}",
                        f"--second={second}",
                        f"--board={board_file}",
                        f"--time={time_inp}",
                    ]

                    try:
                        # Execute the command
                        subprocess.run(command, check=True)
                        print(
                            f"Simulation completed: time={time_inp}, board={board}, first={first}, second={second}"
                        )
                    except subprocess.CalledProcessError as e:
                        print(
                            f"Error during simulation: time={time_inp}, board={board}, first={first}, second={second}"
                        )
                        print(e)
```

```python
import subprocess
import os

for turn in range(0, 10):
    our_player = "team03_A3_best_performing"
    enemy_players = [
        "A3_MCMC",
        "A3_MCTS_with_minimax",
        "greedy_player",
        "team03_A2",
    ]
    for player in enemy_players:
        for first in [our_player, player]:
            if first == our_player:
                second = player
            else:
                second = our_player
            for board in ["2x2", "2x3", "3x2", "3x3", "3x4", "4x4"]:
                board_file = "boards/empty-{board}.txt"
                for time_inp in [0.0, 0.5, 1.0, 5.0]:
                    command = [
                        "python3",  # Or just "python" if using Windows
                        "statistics_simulate_game.py",
                        f"--first={first}",
                        f"--second={second}",
                        f"--board={board_file}",
                        f"--time={time_inp}",
                    ]
                    try:
                        # Execute the command
                        subprocess.run(command, check=True)
                        print(
                            f"Simulation completed: time={time_inp}, board={board}, first={first}, second={second}"
                        )
                    except subprocess.CalledProcessError as e:
                        print(
                            f"Error during simulation: time={time_inp}, board={board}, first={first}, second={second}"
                        )
                        print(e)
```
